### 03.1 Mixed-type Data

```python
# Checking for mixed-type data

for col in customers.columns.tolist():
    weird = (customers[[col]].applymap(type) 
            !=customers[[col]].iloc[0].apply(type)).any(axis=1)
    if len (customers[weird]) > 0:
        print(col)
```

### 03.2 Missing Values

```python
# Checking customers data for missing values

customers.isnull().sum()
```

```python
# Creating a subset of rows missing a 'first_name' value

df_nan = customers[customers['first_name'].isnull() == True]
```

```python
df_nan
```

```python
# Creating a subset of customers data excluding 'first_name' column

customers_clean = customers.drop(columns = ['first_name'])
```

```python
# Creating a subset with only non-missing values for now
# might drop column eventually or replace with Mr./Ms.
# will continue using customers df that includes missing values

customers_no_nan = customers[customers['first_name'].isnull() == False]
```

```python
customers_no_nan.isnull().sum()
```

#### Since the only missing values are in the 'first_name' column and are not particularly required for analysis since 'user_id' is the key column, I've chosen to not alter them. This row may be able to be dropped later in the process.
