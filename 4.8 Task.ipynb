### 6. Create a spending flag for each user based on the average price across all their orders 

```python
# Creating an 'avg_prod_price' column to derive the spending flag from

ords_prods_merge['avg_prod_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
# Flagging users with avg_prod_price < 10 as 'Low spenders'

ords_prods_merge.loc[ords_prods_merge['avg_prod_price'] < 10,
        'spending_flag'] = 'Low spender'
```

```python
# Flagging users with avg_prod_price >= 10 as 'High spenders'

ords_prods_merge.loc[(ords_prods_merge['avg_prod_price'] >= 10),
         'spending_flag'] = 'High spender'
```

```python
# Checking the number of high and low spenders

ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

### 7. Create an order frequency flag that marks the regularity of a user's ordering behavior according to the median in the 'day_since_prior_order' column

```python
# Creating an 'med_days_since_prior_order' column to derive the regularity flag from

ords_prods_merge['med_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
# Flagging users with med_days_since_prior_order > 20 as 'Non-frequent customer'

ords_prods_merge.loc[ords_prods_merge['med_days_since_prior_order'] > 20,
        'regularity_flag'] = 'Non-frequent customer'
```

```python
# Flagging users with med_days_since_prior_order > 10 & <= 20 as 'Regular customer'

ords_prods_merge.loc[(ords_prods_merge
        ['med_days_since_prior_order'] > 10) & (ords_prods_merge['med_days_since_prior_order'] <= 20),
         'regularity_flag'] = 'Regular customer'
```

```python
# Flagging users with med_days_since_prior_order <= 10 as 'Frequent customer'

ords_prods_merge.loc[ords_prods_merge['med_days_since_prior_order'] <= 10,
        'regularity_flag'] = 'Frequent customer'
```

```python
# Checking the frequency of values in regularity_flag column

ords_prods_merge['regularity_flag'].value_counts(dropna = False)
```
