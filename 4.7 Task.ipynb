# 03. 4.7 Exercise: Deriving Variables

```python
# Creating a subset of the first million rows
df = ords_prods_merge[:1000000]
```

```python
# Creating a price labeling function
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <=15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
# Using the price labeling function on df
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Finding value frequencies in 'price_range'
df['price_range'].value_counts(dropna = False)
```

```python
# Finding maximum price
df['prices'].max()
```

# 04. Task - Deriving New Variables

### 2. Finding the busiest & least busy days of the week

```python
# Creating a function to label the busiest days of the week

result_2 = []

for value in ords_prods_merge["order_day_of_week"]:
    if value == 0 or value == 1 :
        result_2.append("Busiest days")
    elif value == 4 or value == 3:
        result_2.append("Least busy days")
    else:
        result_2.append("Regularly busy days")
```

```python
# Adding a new column for the function

ords_prods_merge['busiest_days_of_week'] = result_2
```

```python
# Checking the value counts of the new column 'busiest_days_of_the_week'

ords_prods_merge['busiest_days_of_week'].value_counts(dropna = False)
```

```python
# Checking the new column 'busiest_days_of_the_week'

ords_prods_merge.head()
```

### 4. Find the busiest hours of the day

```python
# Determining the most and least busy hours of the day

ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Creating a function to label the most and least busy hours of the day

result_3 = []

for value in ords_prods_merge["order_hour_of_day"]:
    if value == 10 or value == 11:
        result_3.append("Most orders")
    elif value == 3 or value == 4:
        result_3.append("Fewest orders")
    else:
        result_3.append("Average orders")
```

```python
# Creating a column for the 'busiest_hours_of_day'

ords_prods_merge['busiest_hours_of_day'] = result_3
```

```python
# Checking the frequency of the values in the new column 'busiest_hours_of_day'

ords_prods_merge['busiest_hours_of_day'].value_counts(dropna = False)
```
